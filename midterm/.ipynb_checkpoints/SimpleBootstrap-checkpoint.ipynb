# Simple Bootstrap

In this notebook you are going to draw samples from a probability distribution of your choice, and then compute a bootstrapped mean and its confidence interval.

## Imports

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

## Boostrapping the mean

Perform the following steps:

* Draw `100` samples from a continuous probability distribution of your choice with a given mean and variance.
* Bootstrap your raw data `500` times and use the bootstrapped data to compute the distribution of the mean of the samples.
* Make an appropriately labeled visualization of the distribution of the mean.
* Compute and print the 90% confidence interval of the mean.

```python
def get_sample_normal(n):
    return np.random.normal(size=n)
```

```python
def bootstrap_data(data):
    dat_len = len(data)
    return np.array([data[ndx] for ndx in np.random.randint(0, dat_len, dat_len)])
```

```python
def get_mean(data):
    return data.mean()
```

```python
def bootstrap_function(data, func, n):
    return np.array([func(bootstrap_data(data)) for i in range(n)])
```

```python
means = bootstrap_function(get_sample_normal(500), get_mean, 500)
```

```python
cf = np.percentile(means, 90)
print("90% Confidence Interval is between: " + str(-cf) + " and " + str(cf))
sns.distplot(means)
sns.axlabel("Means Of Bootstrapped Data", "Count")
```
