## Line plot of sunspot data

Download the `.txt` data for the "Yearly mean total sunspot number [1700 - now]" from the [SILSO](http://www.sidc.be/silso/datafiles) website. Upload the file to the same directory as this notebook.

```python
import os
assert os.path.isfile('yearssn.dat')
```

Use `np.loadtxt` to read the data into a NumPy array called `data`. Then create two new 1d NumPy arrays named `years` and `ssc` that have the sequence of year and sunspot counts.

```python
data = np.loadtxt('yearssn.dat')
year = np.array(data[0::, 0])
ssc = np.array(data[0::, 1])
```

```python
assert len(year)==315
assert year.dtype==np.dtype(float)
assert len(ssc)==315
assert ssc.dtype==np.dtype(float)
```

Make a line plot showing the sunspot count as a function of year.

* Customize your plot to follow Tufte's principles of visualizations.
* Adjust the aspect ratio/size so that the steepest slope in your plot is *approximately* 1.
* Customize the box, grid, spines and ticks to match the requirements of this data.

```python
plt.figure(figsize=(50, 10))
plt.plot(year, ssc)
plt.xlim(1700, 2015)
plt.xticks([x for x in range(1700, 2015, 10)])
plt.ylim(0, 275)
plt.yticks([x for x in range(0, 275, 10)])
plt.tick_params(axis="x", direction="inout", length=10)
plt.tick_params(axis="y", direction="inout", length=10)
plt.grid(True)

plt.title("Average Number of Sun Spots Per Year Since 1700")
plt.xlabel("Year")
plt.ylabel("Average Number of Sun Spots")
```

Describe the choices you have made in building this visualization and how they make it effective.

I did not quite understand the approximate slope of 1 for each line. I elongated the graph to try to make the data more readable, although this means that the labels are much smaller. I decided to set tick marks at ever 10 years for the x axis and every 10 counts on the y axis so that the graph is easier to read and more data can be accummulated by just looking at it. I also added a grid so that the graph is much easier to read. I added titles to the graph and its axes so that the information provided can be read by someone who does not have previous knowledge of the intention of this graph.

Now make 4 subplots, one for each century in the data set. This approach works well for this dataset as it allows you to maintain mild slopes while limiting the overall width of the visualization. Perform similar customizations as above:

* Customize your plot to follow Tufte's principles of visualizations.
* Adjust the aspect ratio/size so that the steepest slope in your plot is *approximately* 1.
* Customize the box, grid, spines and ticks to match the requirements of this data.

```python
f, ax = plt.subplots(2, 2)
f.set_figheight(7)
f.set_figwidth(20)
start_ndx = 0
end_ndx = 100

for i in range(2):
    for j in range(2):
        plt.sca(ax[i, j])
        plt.grid(True)
        plt.plot(year[start_ndx:end_ndx], ssc[start_ndx:end_ndx])
        plt.xlabel("Year")
        plt.ylabel("Average # of Sun Spots")
        plt.tick_params(axis="y", direction="out", length=5)
        plt.tick_params(axis="x", direction="out", length=5)
        
        start_ndx += 100
        end_ndx += 100

plt.tight_layout()
```
